```python
# Most essential modules:

import pandas as pd   # includes tools used in reading data
import numpy as np   # includes tools for numerical calculus
import matplotlib.pyplot as plt  # includes tools used in plotting data

# Other useful modules:

import random as rand   # includes functions in generating random data
from scipy import stats    # includes tools for statistical analysis
from scipy.stats import norm    # tools for normal distribution
import matplotlib.mlab as mlab  # more plotting tools for more complicated diagrams

# Not a module, but essential command which makes the output look prettier in notebooks
%matplotlib inline
```

```python
dimuon = pd.read_table('https://raw.githubusercontent.com/QuarkNet-HEP/2023-ND-QuarkNet-CMS-Data/main/MuRun2010B.csv', sep=',', encoding = "ISO-8859-1")
```

```python
dimuon.head()
```

```python
mass_calc=np.sqrt((dimuon.E1+dimuon.E2)**2-(dimuon.px1+dimuon.px2)**2-(dimuon.py1+dimuon.py2)**2-(dimuon.pz1+dimuon.pz2)**2)
```

```python
print(mass_calc)
```

```python
mass_diff_squd=dimuon.M**2-mass_calc**2
```

```python
print(mass_diff_squd)
```

```python
plt.hist(mass_calc, bins=1000, range=(0,100))
plt.show()
```

```python
plt.xlabel('Invariant mass [GeV]')
plt.ylabel('Number of events')
plt.title('The histogram of the calculated invariant masses of two muons \n')

plt.hist(mass_calc, bins=1000, range=(0,5))
plt.show()
```

```python
plt.xlabel('Invariant mass [GeV]')
plt.ylabel('Number of events')
plt.title('The histogram of the calculated invariant masses of two muons \n')

plt.hist(mass_calc, bins=1000, range=(7.5,12.5))
plt.show()
```

```python
plt.xlabel('Difference Mass**2 from CMS and Mass**2 Calc [GeV]**2')
plt.ylabel('Number of events')
plt.title('The histogram of difference Mass**2 from CMS and Mass**2 Calc [GeV]**2 \n')

plt.hist(mass_diff_squd, bins=1000, range=(-.01,.01))
plt.show()
```

```python
x = mass_diff_squd;
y = dimuon.E1+dimuon.E2;

plt.scatter(x,y)
```

```python
fig, ax = plt.subplots()
for color in ['tab:blue']:
    n = len(dimuon)
    scale = 10
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.02, edgecolors='none')
plt.title('Event total Energy vs difference Mass**2 from CMS and Mass**2 Calc [GeV]**2 \n')
 
plt.xlim([-.01, .01])
plt.ylim([0, 100])
plt.show()
```
